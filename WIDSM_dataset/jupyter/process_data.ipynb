```python
time_segment = [] 
```

```python
ACTIVITY = {
    'Jogging' : 0,
    'Sitting' : 1,
    'Standing' : 2,
    'Walking' : 3,
    'Upstairs' : 4,
    'Downstairs' : 5
}
```

```python
prev_activity = 'No action'
prev_user = 0
```

```python
current_activity = []

for index, row in df.iterrows():
    if index != 0 and (row['activity'] != prev_activity or row['user'] != prev_user):
        time_segment.append(current_activity)
        current_activity = []
    prev_activity = row['activity']
    prev_user = row['user']
    current_activity.append((row['x-axis'], row['y-axis'], row['z-axis'], row['user'], ACTIVITY[row['activity']]))
time_segment.append(current_activity)
    
```

```python
def sliding_window(time_seg, window_size, step, nclass = 6):
    windows = []
    labels_one_hot = []
    left_side = 0
    right_side = window_size
    
    while right_side < len(time_seg):
        this_window = []
        this_label = [0 for i in range(0, nclass)]
        this_label[time_seg[0][4]] = 1
        
        for i in range(left_side, right_side):
            this_window.append(time_seg[i][0:3])
        
        # Slide
        left_side += window_size
        right_side += window_size
        
        
        windows.append(this_window)
        labels_one_hot.append(this_label)
        
    return windows, labels_one_hot
        
    
```

```python
x_data_subjects = [[] for i in range(36)]
y_data_subjects = [[] for i in range(36)]
```

```python
for time_seg in time_segment:
    windows, labels = sliding_window(time_seg, 256, 64)
    subject = time_seg[0][3]
    for e in windows: x_data_subjects[subject - 1].append(e)
    for e in labels: y_data_subjects[subject - 1].append(e)
        
```

```python
for i in range(36):
    print(len(x_data_subjects[i]))
    print(len(y_data_subjects[i]))
```

```python
for i in range(36):
    x_data_subjects[i] = np.array(x_data_subjects[i])
    y_data_subjects[i] = np.array(y_data_subjects[i])
```
